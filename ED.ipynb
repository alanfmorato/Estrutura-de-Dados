#Jornalista quer saber, daquela região, a cidade com mais eleitores jovem e mais eleitores idosos

```python
jovens = {}
for zona in df.NR_ZONA.unique():
    result = df.query('NR_ZONA == @zona').groupby('DS_FAIXA_ETARIA')['QT_ELEITORES_PERFIL'].sum()
    jovens[zona] = result[1:8].sum()
for zona in sorted(jovens, key=lambda x:jovens[x], reverse = True):
    print (f'Zona eleitoral {zona}: {jovens[zona]}')
```

```python
idosos = {}
for zona in df.NR_ZONA.unique():
    result = df.query('NR_ZONA == @zona').groupby('DS_FAIXA_ETARIA')['QT_ELEITORES_PERFIL'].sum()
    idosos[zona] = result[14:22].sum() + int(result[0])
for zona in sorted(idosos, key=lambda x:idosos[x], reverse = True):
    print (f'Zona eleitoral {zona}: {idosos[zona]}')
```

#relacionar as características do eleitorado de acordo com o representante eleito em determinada região

```python
import pandas as pd
from IPython import get_ipython
from IPython.display import display


try:
    df = pd.read_csv('/content/perfil_eleitor_secao_ATUAL_AM.csv',
                     sep=";", on_bad_lines = 'skip', encoding='Latin 1')

    df2 = pd.read_csv('/content/bweb_1t_AM_051020221321.csv',
                     sep=";", on_bad_lines = 'skip', encoding='Latin 1')

    zona_characteristics = df.groupby('NR_ZONA').agg(
        total_eleitores=('QT_ELEITORES_PERFIL', 'sum'),
        jovens=('DS_FAIXA_ETARIA', lambda x: (x.isin(['18 anos', '19 anos', '20 anos', '21 a 24 anos', '25 a 29 anos', '30 a 34 anos', '35 a 39 anos'])).sum()),
        idosos=('DS_FAIXA_ETARIA', lambda x: (x.isin(['60 a 64 anos', '65 a 69 anos', '70 a 74 anos', '75 a 79 anos', '80 a 84 anos', '85 a 89 anos', '90 a 94 anos', '95 a 99 anos', 'Superior a 100 anos'])).sum())
    ).reset_index()

    votacao_raw = {}
    for zona in df2.NR_ZONA.unique():
        result = df2.query('NR_ZONA == @zona').groupby('NM_VOTAVEL')['QT_VOTOS'].sum()
        total = sum(result)
        bolsonaro_votes = result.get("JAIR BOLSONARO", 0)
        lula_votes = result.get("LULA", 0)
        simone_votes = result.get('SIMONE TEBET', 0)
        nulo_votes = result.get('Nulo', 0)
        branco_votes = result.get('Branco', 0)

        votacao_raw[zona] = {
            'Jair Bolsonaro': round(bolsonaro_votes / total * 100, 1) if total > 0 else 0,
            'Lula': round(lula_votes / total * 100, 1) if total > 0 else 0,
            'Simone Tebet': round(simone_votes / total * 100, 1) if total > 0 else 0,
            'Nulos': round(nulo_votes / total * 100, 1) if total > 0 else 0,
            'Brancos': round(branco_votes / total * 100, 1) if total > 0 else 0
        }

    votacao = pd.DataFrame(votacao_raw).transpose().reset_index()
    votacao = votacao.rename(columns={'index': 'NR_ZONA'})

    merged_data = pd.merge(zona_characteristics, votacao, on='NR_ZONA')

    candidatos = ['Jair Bolsonaro', 'Lula', 'Simone Tebet']
    merged_data['Vencedor'] = merged_data[candidatos].idxmax(axis=1)

    print("Dados mesclados com o vencedor por zona:")
    display(merged_data.head())

except FileNotFoundError:
    print("Certifique-se de que os arquivos CSV estejam nos caminhos corretos.")
except KeyError as e:
    print(f"KeyError: {e}. Verifique se as colunas necessárias existem nos DataFrames.")
except Exception as e:
    print(f"Ocorreu um erro: {e}")
```
